```python
base['Height'] = base['Height'].astype('float64')
base['Weight'] = base['Weight'].astype('float64')
base['CH2O'] = base['CH2O'].astype('float64')
base['TUE'] = base['TUE'].astype('float64')
base['NCP'] = base['NCP'].astype('float64')
```

### Put all float numbers with 0.00

```python
def formatar(valor):
  return "{:,.2f}".format(valor)
#---------------------------------------------
base['Height'] = base['Height'].apply(formatar)
base['Weight'] = base['Weight'].apply(formatar)
base['CH2O'] = base['CH2O'].apply(formatar)
base['TUE'] = base['TUE'].apply(formatar)
base['NCP'] = base['NCP'].apply(formatar)
```

## Change file from data frame to data serie

```python
base_series = base.iloc[:,0:17].values
```

```python
base_series[0]
```

## Treatment categorical values - Label Enconder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder_gender = LabelEncoder()
label_encoder_history = LabelEncoder()
label_encoder_FAVC = LabelEncoder()
label_encoder_CAEC = LabelEncoder()
label_encoder_smoke = LabelEncoder()
label_encoder_SCC = LabelEncoder()
label_encoder_CALC = LabelEncoder()
label_encoder_MTRANS = LabelEncoder()
```

```python
base_series[:,0] = label_encoder_gender.fit_transform(base_series[:,0])
base_series[:,4] = label_encoder_history.fit_transform(base_series[:,4])
base_series[:,5] = label_encoder_FAVC.fit_transform(base_series[:,5])
base_series[:,8] = label_encoder_CAEC.fit_transform(base_series[:,8])
base_series[:,9] = label_encoder_smoke.fit_transform(base_series[:,9])
base_series[:,11] = label_encoder_SCC.fit_transform(base_series[:,11])
base_series[:,14] = label_encoder_CALC.fit_transform(base_series[:,14])
base_series[:,15] = label_encoder_MTRANS.fit_transform(base_series[:,15])
```

```python
with open("base_csv.csv", 'w', newline = '') as archive:
  write = csv.writer(archive)
  count = 0
  for arrays in base_series:
    count = 0
    for values in arrays:
      if count == 16:
        print (f"{values}")
      else:
        print (f"{values},",end='')
      count += 1
```
